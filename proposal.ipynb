```r
library(tidyverse)
```

```r
bankdata <- read_csv2("data/bank_additional_full.csv") 
bankdata
```

```r
bankdata_tidy <- select(bankdata,age,duration,campaign,y)
bankdata_tidy
```

```r
deposite_or_not <- bankdata_tidy %>%
group_by(y) %>%
summarize(age_average = mean(age, na.rm = FALSE),
          duration_average = mean(duration, na.rm = FALSE),
          campaign_average = mean(campaign, na.rm = FALSE))
deposite_or_not
```

```r
summary_yes <- bankdata_tidy %>%
            filter( y == "yes")  %>%
            group_by(y)  %>%
            summarize(n = n())

summary_no <- bankdata_tidy %>%
            filter( y == "no")  %>%
            group_by(y)  %>%
            summarize(n = n())

summary_n <- merge(summary_no, summary_yes,  all = T)
summary_n
```

```r
summary <- merge(deposite_or_not,summary_n,all = T)
summary
```
