```r
library(tidyverse)
```

```r
# read date into the jupter
bankdata <- read_csv2("data/bank_additional_full.csv") 
bankdata
```

```r
# select the data that we need and name it as the tidy dataset
bankdata_tidy <- select(bankdata,age,duration,campaign,y)
bankdata_tidy
```

```r
#form a table that contains age average, duration average, campaign averag and count of each class
summary <- bankdata_tidy %>%
group_by(y) %>%
summarize(age_average = mean(age, na.rm = FALSE),
          duration_average = mean(duration, na.rm = FALSE),
          campaign_average = mean(campaign, na.rm = FALSE),
           count = n())
summary
```

```r
# This is histogram plot showing distribution of "duration" predictor on counts of whether the client deposit or not
options(repr.plot.width = 20, repr.plot.height = 10) 
bar_plot_duration <- bankdata_tidy %>%
    ggplot(aes(x = duration, fill = y)) + 
   geom_histogram(position = "identity", alpha = 0.5) +
    xlab("duration") +
    ylab("count") +
    labs(fill = "Does the person deposite") +
    ggtitle("deposite proportion in different duration")+
theme(text = element_text(size = 30))
bar_plot_duration

```

```r
# This is histogram plot showing distribution of "age" predictor on counts of whether the client deposit or not
options(repr.plot.width = 20, repr.plot.height = 10) 
bar_plot_age <- bankdata_tidy %>%
    ggplot(aes(x = age, fill = y)) + 
   geom_histogram(position = "identity", alpha = 0.5) +
    xlab("age") +
    ylab("count") +
    labs(fill = "Does the person deposite") +
    ggtitle("deposite proportion in different age")+
theme(text = element_text(size = 30))
bar_plot_age
```

```r
# This is histogram plot showing distribution of "campaign" predictor on counts of whether the client deposit or not
options(repr.plot.width = 20, repr.plot.height = 10) 
bar_plot_campaign <- bankdata_tidy %>%
    ggplot(aes(x = campaign, fill = y)) + 
   geom_histogram(position = "identity", alpha = 0.5) +
    xlab("campaign") +
    ylab("count") +
    labs(fill = "Does the person deposite") +
    ggtitle("deposite proportion in different campaign")+
    xlim(0,30)+
    ylim(0,17000)+
theme(text = element_text(size = 30))
bar_plot_campaign
```
